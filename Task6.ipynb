1.Load the necessary package for plotting using pyplot from matplotlib. Example - Days(x-axis) represents 8 days and Speed represents a car’s speed. Plot a Basic line plot between days and car speed, put x axis label as days and y axis label as car speed and put title Car Speed Measurement.
         Days=[1,2,3,4,5,6,7,8]

         Speed=[60,62,61,58,56,57,46,63]​​​​​

```python
import matplotlib.pyplot as plt
Day=[1,2,3,4,5,6,7,8]
Speed=[60,62,61,58,56,57,46,63]
plt.plot(Day,Speed,color='g')
plt.xlabel('Days')
plt.ylabel('Car Speed')
plt.title('Car Speed Measurements')
plt.show()
```

4.Plotting a basic sine graph by adding more features. Adding Multiple plots by Superimposition like cosine wave.

```python
import numpy as np
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,color='b',marker='o')
plt.plot(x,z,color='g',marker='*')
plt.xlabel('sine and cosine values from 0 to 4pi')
plt.ylabel('sin(x) and cos(x)')
plt.title('Sine and Cosine Wave')
plt.legend(['sin(x)','cos(x)'])
plt.show()
```

Q.5 Plot Simple bar chart showing popularity of Programming Languages.

Languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']

Popularity = [56, 39, 34, 34, 29]

Security = [44 ,36 ,55, 50, 42]

Plot Multiple Bars showing Popularity and Security of major Programming Languages. Also Create Horizontal bar chart using barh function.

```python
Languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']

Popularity = [56, 39, 34, 34, 29]

Security = [44 ,36 ,55, 50, 42]

plt.subplot(2,1,1)

plt.bar(Languages,Popularity,width=0.5,color='blue',align='center')
plt.xlabel('Language')
plt.ylabel('Popularity')
plt.title("Programming Languages")
plt.legend(['Popularity'])
plt.show()

plt.subplot(2,1,2)

plt.bar(Languages,Security,width=0.3,color='g',align='center')
plt.xlabel('Language')
plt.ylabel('Security')
plt.legend(['Security'])
plt.show()
```

#Horizontal Bar

```python
Languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']

Popularity = [56, 39, 34, 34, 29]

Security = [44 ,36 ,55, 50, 42]

plt.subplot(2,1,1)

plt.barh(Languages,Popularity,color='blue',align='center')
plt.xlabel('Language')
plt.ylabel('Popularity')
plt.title("Programming Languages")
plt.legend(['Popularity'])
plt.show()

plt.subplot(2,1,2)

plt.barh(Languages,Security,color='g',align='center')
plt.xlabel('Language')
plt.ylabel('Security')
plt.legend(['Security'])
plt.show()
```

Q.6 Plot Histogram, We have a sample data of Students marks of various Students, we will try to plot number of Students by marks range and try to figure out how many Students are average, below-average and Excellent.

Marks = [ 61,86,42,46,73,95,65,78,53,92,55,69,70,49,72,86,64]

Histogram showing Below Average, Average and Execellent distribution

40-60: Below Average

60-80: Average

80-100: Excellent

```python
Marks=np.array([61,86,42,46,73,95,65,78,53,92,55,69,70,49,72,86,64])
below_avg=Marks[np.logical_and(Marks>=40,Marks<60)]
avg_marks=Marks[np.logical_and(Marks>=60,Marks<80)]
exc_marks=Marks[np.logical_and(Marks>=80,Marks<=100)]
plt.hist(Marks)
plt.show()
print("Below Average Students are:",below_avg.size)
print("Average Students are :",avg_marks.size)
print("Excellent Students are :",exc_marks.size)
```

Q.7 Titanic Data Set Download Data

 Load the data file

 (i) Create a pie chart presenting the male/female proportion 

 (ii) Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
import pandas as pd
titanic=pd.read_csv("titanic_original.csv")
titanic.head()
```

```python
titanic['sex']
x=titanic.sex.value_counts()
x
```

```python
plt.pie(x, autopct='%.0f%%')

plt.legend(['Male','Female'])

plt.title('Male And Female Proportion')

plt.show()
```

```python
male = titanic[titanic['sex']=='male']

male.sex.value_counts()

female = titanic[titanic['sex']=='female']

female.sex.value_counts()

plt.scatter(male.fare, male.age, color='g',label = 'Male', marker = '^')

plt.scatter(female.fare, female.age, color='orange',label = 'Female', marker = '*')

plt.legend()

plt.xlabel('fare')

plt.ylabel('age')

plt.grid()

plt.show()
```
